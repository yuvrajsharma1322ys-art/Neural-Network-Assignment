**Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.**

Answer:

Deep Learning is a subset of Machine Learning that is inspired by the structure of the human brain (neurons). It uses neural networks with multiple layers that automatically learn features from data.

Evolution: In the 1950s, the perceptron model was introduced. In the 1980s, backpropagation was developed, which made it possible to train multilayer networks. However, progress was slow due to limited computing power. After 2012, with the availability of GPUs and large datasets, Deep Learning grew rapidly (CNNs, RNNs, Transformers, etc).

Difference from traditional ML:
- In traditional ML, features need to be manually engineered.
- Deep Learning automatically extracts features from raw data using multiple hidden layers.
- Deep Learning requires more data and computing power, while traditional ML can perform well with smaller datasets.
- Traditional ML uses simple algorithms (Linear Regression, Decision Tree), while Deep Learning uses neural networks.

**Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?**

Answer:

A Perceptron is the simplest type of neural network, consisting of only a single layer. It is mainly used for binary classification.

Architecture:
- Inputs (x1, x2, x3...) are received, each multiplied with corresponding weights (w1, w2, w3...).
- The weighted inputs are summed and a bias is added: z = (w1*x1 + w2*x2 + ... ) + b
- This sum is passed through an activation function (such as a step function) which produces an output of 0 or 1.

Working: During training, the perceptron adjusts its weights so that predictions become correct. If a prediction is wrong, the weights are updated based on the error.

Limitations:
- A perceptron can only solve linearly separable problems, like the AND or OR gate.
- It cannot solve non-linear problems like XOR.
- Because it has only a single layer, it cannot learn complex patterns.
- This is why the multilayer perceptron (MLP) with hidden layers was introduced.

**Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.**

Answer:

The purpose of an activation function is to add non-linearity to the network. Without an activation function, a neural network would behave just like a linear equation, no matter how many layers it has. The activation function is what allows the network to learn complex patterns.

Comparison:

1. Sigmoid:
   - Formula: 1 / (1 + e^(-x))
   - Output range is between 0 and 1.
   - Good for probability-based output (binary classification).
   - Problem: vanishing gradient, which causes slow learning in deep networks.

2. ReLU (Rectified Linear Unit):
   - Formula: max(0, x)
   - Output is either 0 or a positive value.
   - Fast and computationally simple, hence it is most widely used in deep networks.
   - Problem: sometimes causes the "dying ReLU" problem where a neuron always outputs 0.

3. Tanh:
   - Formula: (e^x - e^-x) / (e^x + e^-x)
   - Output range is between -1 and 1.
   - Better than sigmoid because it is zero-centered.
   - Problem: also faces vanishing gradient issues in deep networks.

**Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.**

Answer:

Loss function and Cost function both measure error, but their scope is different.

Loss Function: This calculates the error for a single training example only.
Example: Mean Squared Error for one sample = (y_actual - y_predicted)^2

Cost Function: This is the average of the loss over the entire dataset (or batch).
Example: Cost = (1/n) * sum of (y_actual - y_predicted)^2 for all n samples

Simple example:
- In a regression problem, the loss function could be Mean Squared Error (for a single example), and the cost function is the average MSE (over all examples).
- In a classification problem, the loss function is Cross Entropy Loss (for a single example), and the cost function is the average cross entropy (over all examples).

In short, loss = error for one data point, cost = average error over all data points, which the model tries to minimize during training.

**Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.**

Answer:

The role of an optimizer is to update the model's weights in order to minimize the loss/cost function. It uses gradients (derivatives) to move the weights in the right direction.

Comparison:

1. Gradient Descent:
   - The most basic optimizer.
   - Calculates the gradient over the entire dataset to update weights.
   - Slow on large datasets, since the whole dataset is used in every step.

2. Adam (Adaptive Moment Estimation):
   - Combines the benefits of both Momentum and RMSprop.
   - Adapts the learning rate for each parameter individually.
   - Gives fast convergence, hence it is the most popular optimizer in deep learning.

3. RMSprop:
   - Adjusts the learning rate based on a moving average of squared gradients.
   - Reduces oscillations and gives faster training compared to plain gradient descent.
   - Widely used in models like RNNs.

Overall: Gradient Descent is simple but slow, while RMSprop and Adam are more efficient and faster for practical deep learning models.

**Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.**

Answer:

```python
import numpy as np

# AND gate input and output data
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

# initialize weights and bias
weights = np.zeros(2)
bias = 0
lr = 0.1
epochs = 10

def step_function(z):
    return 1 if z >= 0 else 0

# training the perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = step_function(z)
        error = y[i] - y_pred
        weights = weights + lr * error * X[i]
        bias = bias + lr * error

print("Final weights:", weights)
print("Final bias:", bias)

# testing the perceptron
print("\nPredictions for AND gate:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    y_pred = step_function(z)
    print(f"Input: {X[i]} -> Predicted: {y_pred}, Actual: {y[i]}")
```

**Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.**

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

plt.figure(figsize=(10,6))

plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, tanh(x), label="Tanh")

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

**Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.**

Answer:

```python
import tensorflow as tf
from tensorflow import keras

# load mnist dataset
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

# normalize data
X_train = X_train / 255.0
X_test = X_test / 255.0

# build simple multilayer neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])

model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

history = model.fit(X_train, y_train, epochs=5, validation_split=0.1)

# print training accuracy
train_loss, train_acc = model.evaluate(X_train, y_train)
print("Training Accuracy:", train_acc)
```

**Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.**

Answer:

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# load fashion mnist dataset
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])

model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

history = model.fit(X_train, y_train, epochs=5, validation_split=0.1)

# plot loss curve
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

# plot accuracy curve
plt.subplot(1,2,2)
plt.plot(history.history["accuracy"], label="Train Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```

Interpretation: As the number of epochs increases, the training loss decreases and the training accuracy increases, which shows that the model is learning correctly. If the validation loss starts increasing much more than the training loss, it indicates that the model is overfitting.

**Question 10: Bank fraud detection project - Real-time data science workflow**

Answer:

This is a binary classification problem (fraud vs legitimate) where the data is imbalanced (fraud cases are much fewer compared to legitimate cases).

1. Model Design:
   - I would use a Multilayer Neural Network (MLP) since the data is structured/tabular.
   - The input layer would have as many neurons as there are features (transaction amount, merchant ID, location etc).
   - I would use 2-3 hidden layers with ReLU activation, and the output layer would have 1 neuron with sigmoid activation (binary output).

2. Activation and Loss function:
   - In the hidden layers, I would use ReLU activation since it is fast and effective.
   - In the output layer, I would use Sigmoid activation since we need a probability output (0 to 1).
   - Loss function: Binary Cross Entropy, since it is best suited for binary classification.

3. Training and evaluation with class imbalance:
   - To handle class imbalance, I would use the class_weight parameter or oversample the minority class (fraud) using the SMOTE technique.
   - I would not rely only on accuracy, since accuracy can be misleading on imbalanced data. Instead I would use Precision, Recall, F1-score, and the AUC-ROC curve.
   - I would split the data into train-test sets using a stratified split so that both classes remain in proportion.

4. Optimizer and overfitting prevention:
   - Optimizer: Adam, since it gives fast convergence and uses an adaptive learning rate.
   - To prevent overfitting, I would use Dropout layers, Early Stopping, and L2 regularization.
   - Cross-validation can also be used to check the model's generalization.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow import keras

# example with dummy data (real project me actual bank dataset use hoga)
np.random.seed(42)
X = np.random.rand(1000, 5)   # 5 features jaise transaction amount, merchant id etc
y = np.random.choice([0,1], size=1000, p=[0.95, 0.05])   # imbalanced data, 5% fraud

# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# build model
model = keras.Sequential([
    keras.layers.Dense(32, activation="relu", input_shape=(5,)),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(16, activation="relu"),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])

# handle class imbalance using class_weight
class_weight = {0: 1, 1: 19}   # giving more weight to fraud class

early_stop = keras.callbacks.EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)

history = model.fit(X_train, y_train, epochs=20, validation_split=0.1,
                     class_weight=class_weight, callbacks=[early_stop])

# evaluate
loss, acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", acc)
```
